```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
%matplotlib inline
```

```python
df = pd.read_csv("C:/Users/UserName/201912_20200317 .csv",encoding="cp949")
target = pd.read_csv("C:/Users/UserName/Downloads/Target_customer.csv",encoding="cp949")
df_join = df.join(target,on="주문자 회원번호",how="left")
df_join = df_join[df_join["정산 주문 유형"] == "반품"]
print(df_join)
df_business =df_join.pivot_table(index="주문자 업태",values = "주문자 회원번호", columns="업체 명",aggfunc='count')
plt.figure(figsize=(12,12))
plt.title("업태 및 협력사별 반품 회원 수",fontdict={'fontsize' : 20})
sns.heatmap(data=df_business,cmap="Blues")
```
